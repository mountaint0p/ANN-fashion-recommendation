```python
import os
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Load the pre-trained VGG16 model
model = VGG16(weights='imagenet')

# Path to the folder containing images
images_folder = 'archive/images'

# Get the first 10 image file paths from the folder
image_files = [os.path.join(images_folder, file) for file in os.listdir(images_folder) if file.lower().endswith(('png', 'jpg', 'jpeg'))][:10]

# Function to preprocess an image for VGG16
def preprocess_image(image_path):
    img = load_img(image_path, target_size=(224, 224))  # Resize the image to 224x224
    img_array = img_to_array(img)  # Convert image to array
    img_array = preprocess_input(img_array)  # Preprocess the image for VGG16
    return tf.expand_dims(img_array, axis=0)  # Add batch dimension

# Loop through the first 10 images
for i, image_path in enumerate(image_files, start=1):
    try:
        # Preprocess the image
        preprocessed_image = preprocess_image(image_path)

        # Predict the image class using VGG16
        predictions = model.predict(preprocessed_image)

        # Decode and print the top-3 predictions
        decoded_predictions = decode_predictions(predictions, top=3)[0]
        print(f"Image {i}: {os.path.basename(image_path)}")
        for rank, (class_id, class_name, score) in enumerate(decoded_predictions, start=1):
            print(f"  {rank}: {class_name} ({score:.2f})")
        print()
    except Exception as e:
        print(f"Error processing image {image_path}: {e}")
```

```python
from tensorflow.keras.applications.vgg16 import VGG16, decode_predictions
import json
import requests

# Load the ImageNet class index
imagenet_url = 'https://storage.googleapis.com/download.tensorflow.org/data/imagenet_class_index.json'
response = requests.get(imagenet_url)
imagenet_class_index = response.json()

# Extract all class labels
vgg16_classes = [imagenet_class_index[str(i)][1] for i in range(len(imagenet_class_index))]

# Print or save the class list
print(f"Total Classes: {len(vgg16_classes)}")
print(vgg16_classes)

# save the class list to a file
with open('vgg16_classes.txt', 'w') as f:
    f.write('\n'.join(vgg16_classes))
```
